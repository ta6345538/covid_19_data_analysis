### Q. 1) Show the number of Confirmed, Deaths and Recovered cases in each Region.

```python
plot_data = data.groupby('Region')['Confirmed'].sum().sort_values(ascending = False).head(5)
plot_data
```

```python
plot_data.plot(kind = 'bar')
```

```python
data.groupby('Region')['Confirmed','Deaths','Recovered'].sum().head()
```

### Q. 2) Remove all the records where the Confirmed Cases is Less Than 10.

```python
data_n = data[~(data.Confirmed < 10)]
data_n.head(10)
```

### Q. 3) In which Region, maximum number of Confirmed cases were recorded ?

```python
data.groupby('Region')['Confirmed'].sum().sort_values(ascending = False).head(15)
```

### Q. 4) In which Region, minimum number of Deaths cases were recorded ?

```python
data.groupby('Region')['Deaths'].sum().sort_values(ascending = True).head(50)
```
